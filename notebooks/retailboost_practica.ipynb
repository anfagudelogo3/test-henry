```python
import pandas as pd

clientes = pd.DataFrame({
    "cliente":          ["Ana", "Luis", "Marta", "Jorge", "Sofía", "Diego"],
    "ingresos":         [3000,  2200,   5000,    1800,    4200,    2600],   # $ que gana al mes
    "gastos":           [2500,  2300,   3100,    1500,    2000,    2800],   # $ que gasta al mes
    "visitas":          [10,    3,      20,      2,       25,      4],      # veces que entró al sitio
    "minutos_en_sitio": [40,    30,     300,     8,       400,     24],     # minutos totales en el sitio
    "productos_vistos": [12,    2,      60,      1,       90,      3],      # productos que miró
})
clientes
```

---

## ✅ Ejemplo resuelto: `ingreso_disponible`

Tener **ingresos altos** no significa que pueda gastar… ¡quizás también gasta mucho!
Lo que realmente importa es **cuánto le sobra**:

> 💰 **ingreso_disponible = ingresos − gastos**

**¿Por qué tiene sentido?** Un cliente al que le sobra plata tiene más capacidad de comprar
que uno que gasta todo lo que gana. La **resta** captura algo que ninguna de las dos columnas dice por separado.

```python
clientes["ingreso_disponible"] = clientes["ingresos"] - clientes["gastos"]
clientes[["cliente", "ingresos", "gastos", "ingreso_disponible"]]
```

> 👀 ¡Mirá! A **Luis** y a **Diego** les da **negativo**: gastan más de lo que ganan.
> Eso es justo lo que un modelo necesita saber. 💡
