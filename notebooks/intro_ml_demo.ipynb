# 🎓 3. Tipos de aprendizaje — ⏱️ 1.5 min

Hay **tres formas de aprender** en ML:

| Tipo | Idea | En RetailBoost 🛒 |
|---|---|---|
| ✅ **Supervisado** | Aprende con **respuestas conocidas** | Clientes viejos de los que **ya sabemos si compraron** |
| 🔍 **No supervisado** | Encuentra **grupos** sin respuestas | Agrupar clientes en perfiles |
| 🎮 **Por refuerzo** | Aprende **probando**, con premios | Probar qué cupón vende más |

El **supervisado** se divide en **regresión** (predice un número: *¿cuánto gastará?*) y **clasificación** (predice una etiqueta: *¿compra SÍ/NO?*).

<div style='background:#e8f5e9; border-left:5px solid #4caf50; padding:12px; border-radius:8px; margin:12px 0;'>

👉 Nuestro caso es **aprendizaje supervisado de clasificación**: ¿este visitante compra, **SÍ o NO**? ¡Construyámoslo! 🙌

# 🧩 4. Elementos de un modelo — ⏱️ 3.5 min

Abramos los datos reales de RetailBoost 👇

```python
# ── Herramientas ─────────────────────────────────────────────────────────────
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, recall_score

clientes = pd.read_csv("../data/raw/retailboost_customers.csv")
print(f"✅ {len(clientes):,} visitantes | {clientes.shape[1]} columnas")
clientes.head()
```

Cada columna es un dato del visitante. Estos son los **ladrillos** de todo modelo 🧱

| Concepto | En simple | En RetailBoost |
|---|---|---|
| 🔢 **Features** | Lo que el modelo **mira** | edad, ingresos, visitas, minutos… |
| 🎯 **Target** | Lo que **predice** | `compro` (1 = sí, 0 = no) |
| 👤 **Muestra** | Cada **fila** | un visitante |
| ⚙️ **Algoritmo** | La **receta** que aprende | Árbol de decisión |
| 🎚️ **Parámetros / Hiperparámetros** | Lo que **aprende solo** / lo que **elegimos** | las preguntas del árbol / su profundidad |

```python
# Los modelos solo entienden NÚMEROS y no toleran vacíos → preparamos mínimamente
datos = clientes.copy()
datos["ingresos_mensuales"] = datos["ingresos_mensuales"].fillna(datos["ingresos_mensuales"].median())
datos["region"] = datos["region"].fillna("Desconocida")
datos["region_cod"] = LabelEncoder().fit_transform(datos["region"])   # texto → número

FEATURES_CRUDAS = ["edad", "ingresos_mensuales", "gasto_mensual",
                   "visitas_mes", "minutos_en_sitio", "productos_vistos", "region_cod"]
TARGET = "compro"

print("🔢 Features:", FEATURES_CRUDAS)
print("🎯 Target  :", TARGET)
```

Separamos los datos: una parte para **aprender** (train) y otra **escondida** para **tomarle examen** (test).
Así sabemos si aprendió de verdad… o solo memorizó. 📝

```python
def entrenar_y_evaluar(df, features, etiqueta):
    X, y = df[features], df[TARGET]
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.25, random_state=42, stratify=y)   # 75% aprende / 25% examen
    modelo = DecisionTreeClassifier(max_depth=4, random_state=42)  # max_depth = hiperparámetro
    modelo.fit(X_train, y_train)                                    # aquí APRENDE
    pred = modelo.predict(X_test)
    acc, rec = accuracy_score(y_test, pred), recall_score(y_test, pred)
    print(f"📊 {etiqueta}")
    print(f"   Variables usadas: {len(features)}")
    print(f"   Accuracy (acierto global)      : {acc:.1%}")
    print(f"   Recall   (compradores hallados): {rec:.1%}")
    return acc, rec

acc_crudas, rec_crudas = entrenar_y_evaluar(datos, FEATURES_CRUDAS, "Modelo v1 — SOLO variables crudas")
```

💬 **Chat:** acierta ~**67%**… ¿alcanza para que la CEO confíe? ¿Cómo **ayudamos al modelo a ver mejor**?

</div>

> ⚖️ *Nota rápida:* el `max_depth` evita dos extremos → **subajuste** (modelo muy simple, no aprende) y **sobreajuste** (memoriza y falla con clientes nuevos).

# 🛠️ 5. Ingeniería de características — ⏱️ 3.5 min ⭐

> **La parte que más mejora un modelo:** crear **variables nuevas** combinando las que ya tenemos.

Mirá dos clientes. Por separado, *"ingresos = 3.000"* dice poco. **Combinando** dice todo 👇

| | Ingresos | Gasto | Visitas | Minutos | → 💰 sobra (ing−gasto) | → ⏱️ min/visita |
|---|---|---|---|---|---|---|
| 👩 Ana   | \$3.000 | \$2.500 | 10 | 40 | **+\$500** | **4** |
| 👨 Diego | \$2.600 | \$2.800 | 4  | 24 | **−\$200** | **6** |

Creemos esas variables nuevas 👇

```python
# 🛠️ Variables NUEVAS combinando las que ya teníamos
datos["ingreso_disponible"]   = datos["ingresos_mensuales"] - datos["gasto_mensual"]   # 💰 cuánto le sobra
datos["minutos_por_visita"]   = datos["minutos_en_sitio"]   / datos["visitas_mes"]     # ⏱️ qué tan enganchado
datos["productos_por_visita"] = datos["productos_vistos"]   / datos["visitas_mes"]     # 👀 qué tan interesado

datos[["ingresos_mensuales", "gasto_mensual", "ingreso_disponible",
       "minutos_en_sitio", "visitas_mes", "minutos_por_visita"]].head()
```

```python
# Mismo modelo, mismas filas... pero sumando las 3 variables que inventamos nosotros
FEATURES_DERIVADAS = ["ingreso_disponible", "minutos_por_visita", "productos_por_visita"]
acc_eng, rec_eng = entrenar_y_evaluar(
    datos, FEATURES_CRUDAS + FEATURES_DERIVADAS, "Modelo v2 — crudas + variables creadas")

print("\n" + "=" * 50)
print("📈 ANTES vs DESPUÉS de la ingeniería de características")
print(f"   Accuracy:  {acc_crudas:.1%}  →  {acc_eng:.1%}   (+{(acc_eng-acc_crudas)*100:.1f} pts)")
print(f"   Recall  :  {rec_crudas:.1%}  →  {rec_eng:.1%}   (+{(rec_eng-rec_crudas)*100:.1f} pts)")
print("=" * 50)
```
